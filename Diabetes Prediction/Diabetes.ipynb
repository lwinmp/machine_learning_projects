```python
X = data.drop(columns='Outcome', axis=1)
y = data['Outcome']
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit(X)
transfered_data = scaler.transform(X)
```

```python
print(scaled_data)
```

```python
print(transfered_data)
```

```python
X = transfered_data
y = data['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
trained_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
train_acc = accuracy_score(trained_pred, y_train)
print(f'Model Prediction accuracy on Training data is {train_acc:.2%}')
```

```python
test_acc = accuracy_score(test_pred, y_test)
print(f'Model Prediction accuracy on Testing data is {test_acc:.2%}')
```
