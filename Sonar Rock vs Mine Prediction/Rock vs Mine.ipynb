```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("sonar data.csv")
data.shape
```

```python
X = data.drop(columns='R', axis=1)
y = data['R']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1) #10% test 90% train
```

```python
print(data.shape)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
train_acc = accuracy_score(train_pred, y_train)
print(f'Model Prediction accuracy on Training data is {train_acc:.2%}')
```

```python
test_acc = accuracy_score(test_pred, y_test)
print(f'Model Prediction accuracy on Testing data is {test_acc:.2%}')
```
